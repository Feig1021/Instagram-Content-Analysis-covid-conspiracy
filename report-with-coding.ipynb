In particular, discussion about COVID-19 origination has snowballed significantly. Through the conspiracist lens, legitimate questions about it become accusations waged against states, especially China, in terms of COVID-19 as a Chinese bioweapon (Birchall, Knight, 2020). The first QAnon post on coronavirus claimed the same argument (QAnon, 2020, March 24).
```python
#import Regx to find all the most related keyword for futher visulise in RankFlow

import re

Filenames = ["Feb_all_textposting.txt", "Mar_all_textposting.txt","Apr_all_textposting.txt", "May_all_textposting.txt"]
patterns = []
freq_list = []

for x in Filenames:
    a = open(x,'r',encoding ='utf-8')
    b = a.read()
    c = b.lower()
    d = re.findall(r"\w+gate|\w+virus|5g\b", c)  # to find all word end with "gate" and "virus", and all "5g" 
    patterns.append(d)

for x in range(len(patterns)):
    y = Counter(patterns[x])
    listwords = [ [k,v] for k, v in y.items() ]
    freq_list.append(listwords)
    
freq_list

```

```python
# the function is to export seperatly export each month data
def virustable(x):
    df = pd.DataFrame(freq_list[x], columns = ["Word", "Counts"])
    y = df.sort_values("Counts", ascending=False)
    virus= y.reset_index(drop= True)
    virus.to_csv('virus.csv', index=False)  # export as csv file 
    return

#start from 0 until 3, and after every output to download the virus.csv file in locoal desktop
virustable(3)
```

```python
# import NRClex to conduct sentiment analysis

from nrclex import NRCLex 

emotion_scores = []

for item in Filenames:
    a = open(item, "r", encoding="utf-8")
    b = a.read()
    emotion = NRCLex(b)
    emotion_scores.append(emotion.raw_emotion_scores)
    
print(emotion_scores)
```

```python
mouth =['February', 'March', 'April', 'May']
df = pd.DataFrame(emotion_scores)
df['Mouth'] = mouth
emotion_class = df.set_index('Mouth')
x = emotion_class.T
x
```

```python
ax = x.plot.bar()
plt.rcParams["figure.figsize"]=15,10
#ax.set_xticks(x.index)
ax.set_xticklabels(x.index, fontsize = 15, rotation=0)
ax.set_title('Emotion Classification', fontsize = 30, color='black')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False) 
plt.rc('legend',fontsize=10) 
ax.yaxis.set_visible(False)

plt.savefig('emotion_classification.png')
```
